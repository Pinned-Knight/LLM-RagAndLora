```python
def getCategory(prompt):
    categories = ["CMS", "E-Commerce", "Listing", "Educational", "Publishing", "None"]
    pre = (
        "You are a classifier.\n"
        "Given a prompt to build a website (enclosed between # symbols), identify its purpose.\n"
        "There are SIX possible categories: CMS, E-Commerce, Listing, Educational, Publishing, None.\n"
        "Assign a confidence score (between 0 and 1) to EACH category, so the sum does NOT have to be 1.\n"
        "Output EXACTLY one line: a Python dictionary with category names as keys and scores as float values.\n"
        "Example format: {\"CMS\":0.72, \"E-Commerce\":0.88, \"Listing\":0.32, \"Educational\":0.01, \"Publishing\":0.25, \"None\":0.00}\n"
        "Instructions:\n"
        "- If the prompt is strongly about one category, give it a high score (e.g., 0.9+) and others very low (e.g., 0.01).\n"
        "- If the prompt is a mix of categories, give high scores to relevant ones.\n"
        "- If the prompt is unrelated to websites, set 'None' to 1.0 and all others to 0.0.\n"
        "Do NOT include any explanation or extra text. Output ONLY the dictionary on one line.\n\n"
        "Examples:\n"
        "# Create a WordPress site to publish news articles #\n"
        "{\"CMS\":0.9, \"E-Commerce\":0.01, \"Listing\":0.01, \"Educational\":0.01, \"Publishing\":0.85, \"None\":0.0}\n"
        "# Build an online store selling crafts #\n"
        "{\"CMS\":0.01, \"E-Commerce\":0.95, \"Listing\":0.01, \"Educational\":0.01, \"Publishing\":0.01, \"None\":0.0}\n"
        "# Directory of local restaurants #\n"
        "{\"CMS\":0.01, \"E-Commerce\":0.01, \"Listing\":0.92, \"Educational\":0.01, \"Publishing\":0.01, \"None\":0.0}\n"
        "# Build a platform for online courses #\n"
        "{\"CMS\":0.01, \"E-Commerce\":0.01, \"Listing\":0.01, \"Educational\":0.93, \"Publishing\":0.01, \"None\":0.0}\n"
        "# This prompt is about cooking recipes #\n"
        "{\"CMS\":0.01, \"E-Commerce\":0.01, \"Listing\":0.01, \"Educational\":0.01, \"Publishing\":0.01, \"None\":1.0}\n"
        "Now classify the following prompt:\n#\n"
    )
    post = "\n#\nEnd of prompt. Respond now using the exact dictionary format."
    output = getResponse(pre + prompt + post)
    # Parse output: expected format is a dictionary string
    try:
        category_scores = eval(output.strip(), {"__builtins__": None}, {})
        if isinstance(category_scores, dict):
            # Ensure all required categories are present
            for cat in categories:
                if cat not in category_scores:
                    category_scores[cat] = 0.0
            # Ensure all values are float
            for k in category_scores:
                try:
                    category_scores[k] = float(category_scores[k])
                except Exception:
                    category_scores[k] = 0.0
            # Only keep required categories
            category_scores = {k: category_scores[k] for k in categories}
            return category_scores
    except Exception:
        pass
    # Fallback in case of parsing error
    fallback = {k: 0.0 for k in categories}
    fallback["None"] = 1.0
    return fallback
```

```python

def extract_pdf_text(pdf_path):
    """
    Extracts and returns all text from a PDF file given its path.
    Args:
        pdf_path (str): The path to the PDF file.
    Returns:
        str: The extracted text from the PDF.
    """
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() or ""
    return text
```

```python
from bs4 import BeautifulSoup

def extract_all_text_for_classification(html_path):
    """
    Extracts as much textual information as possible from an HTML file for LLM classification.
    Includes visible text, meta tags (title, description, keywords), and image alt text.
    Args:
        html_path (str): Path to the HTML file.
    Returns:
        str: Concatenated text content for classification.
    """
    with open(html_path, 'r', encoding='utf-8') as f:
        html = f.read()
    soup = BeautifulSoup(html, 'html.parser')

    # Meta tags: title, description, keywords
    meta_texts = []
    if soup.title and soup.title.string:
        meta_texts.append(soup.title.string)
    for meta in soup.find_all('meta'):
        if meta.get('name', '').lower() in ['description', 'keywords']:
            content = meta.get('content')
            if content:
                meta_texts.append(content)

    # Image alt texts
    alt_texts = [img.get('alt', '') for img in soup.find_all('img') if img.get('alt')]

    # Remove script, style, and irrelevant tags
    for tag in soup(['script', 'style', 'noscript', 'footer', 'nav', 'aside']):
        tag.decompose()

    # Visible text from main, article, or body
    main_content = soup.find(['main', 'article']) or soup.body
    visible_text = main_content.get_text(separator="\n", strip=True) if main_content else ""

    # Combine all text
    all_text = "\n".join(meta_texts + alt_texts + [visible_text])
    return all_text

# Example usage:
# text = extract_all_text_for_classification('Websites/Joomla.html')
# print(text)
```
